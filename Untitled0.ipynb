```python
def remove_zeroes(numbers: list[int]) -> int:
    non_zero_index = 0

    # Перебираем элементы списка
    for i in range(len(numbers)):
        if numbers[i] != 0:
            # Если текущий элемент не равен нулю,
            # перемещаем его в начало списка
            numbers[non_zero_index] = numbers[i]
            non_zero_index += 1

    # Заполняем оставшуюся часть списка нулями
    for i in range(non_zero_index, len(numbers)):
        numbers[i] = 0

    return non_zero_index

# Пример использования
nums = [0, 42, 21, 0, 100, 0, 5, 1, 0, 7, 3, 0, 404, 0]
new_len = remove_zeroes(nums)

assert new_len == 8
assert nums[:new_len] == [42, 21, 100, 5, 1, 7, 3, 404]

def move_zeros_to_end(arr):
    zero_pos = len(arr) - 1
    for i in range(len(arr) - 1, -1, -1):
        if arr[i] == 0:
            arr.pop(i)
            arr.append(0)
    return arr

def test_move_zeros_to_end():
    # Тест на обычном массиве с нулями
    input_array = [0, 1, 0, 3, 12]
    expected_output = [1, 3, 12, 0, 0]
    assert move_zeros_to_end(input_array) == expected_output

    # Тест на массиве без нулей
    input_array = [1, 2, 3, 4, 5]
    expected_output = [1, 2, 3, 4, 5]
    assert move_zeros_to_end(input_array) == expected_output

    # Тест на массиве, где все элементы - нули
    input_array = [0, 0, 0, 0, 0]
    expected_output = [0, 0, 0, 0, 0]
    assert move_zeros_to_end(input_array) == expected_output

    # Тест на пустом массиве
    input_array = []
    expected_output = []
    assert move_zeros_to_end(input_array) == expected_output

    # Тест на массиве, где только один ноль
    input_array = [0]
    expected_output = [0]
    assert move_zeros_to_end(input_array) == expected_output

    # Тест на массиве, где нули и другие элементы чередуются
    input_array = [1, 0, 2, 0, 3, 0, 4, 0, 5]
    expected_output = [1, 2, 3, 4, 5, 0, 0, 0, 0]
    assert move_zeros_to_end(input_array) == expected_output

    print("Все тесты пройдены успешно!")

# Запуск теста
test_move_zeros_to_end()


```

URL-кодирование

```python
def urlify(text: list[str]) -> int:
    # Инициализация указателей для перемещения по массиву
    slow = fast = 0
    # Счетчик пробелов в исходном тексте
    count_space = 0

    # Первый проход: подсчет пробелов
    while fast < len(text):
        # Прекращаем итерацию, если достигнут конец текста или пустой символ
        if text[fast] == '':
            break
        # Увеличиваем счетчик пробелов, если встречаем пробел
        if text[fast] == ' ':
            count_space += 1
        fast += 1

    # Удаляем все элементы после последнего непустого символа
    del text[:fast - 1:-1]
    # Вычисляем новый размер массива после замены пробелов на '%20'
    new_size = fast - 1 + (2 * count_space)
    # Добавляем нужное количество пустых элементов для хранения '%20'
    text += [''] * (2 * count_space)

    # Второй проход: замена пробелов на '%20'
    slow = len(text) - 1
    fast -= 1

    while fast >= 0:
        if text[fast] == ' ':
            # Заменяем пробел на '%20'
            text[slow] = '0'
            text[slow - 1] = '2'
            text[slow - 2] = '%'
            # Сдвигаем указатель slow на 3 позиции назад
            slow -= 3
        else:
            # Копируем символ из исходной позиции в новую позицию
            text[slow] = text[fast]
            slow -= 1

        fast -= 1

    # Возвращаем новый размер массива после замены
    return new_size + 1


# Примеры тестов:

# Пример 1: обычный текст с пробелами
text = list('my url') + [''] * 21
new_len = urlify(text)
assert new_len == 8
assert text[:new_len] == list('my%20url')

# Пример 2: пустая строка
text = list('')
new_len = urlify(text)
assert new_len == 0
assert text[:new_len] == list('')

# Пример 3: текст с пробелами в начале и в конце
text = list(' my url ')
new_len = urlify(text)
assert new_len == 14
assert text[:new_len] == list('%20my%20url%20')

# Пример 4: текст с несколькими пробелами между словами
text = list(' my   url ')
new_len = urlify(text)
assert new_len == 20
assert text[:new_len] == list('%20my%20%20%20url%20')

# Пример 5: текст с пробелами и словами
text = list(' my url with multiple spaces and words ')
new_len = urlify(text)
assert new_len == 55
assert text[:new_len] == list('%20my%20url%20with%20multiple%20spaces%20and%20words%20')

# Пример 6: текст с одним пробелом
text = list(' ')
new_len = urlify(text)
assert new_len == 3
assert text[:new_len] == list('%20')

# Пример 7: текст без пробелов
text = list('myurl')
new_len = urlify(text)
assert new_len == 5
assert text[:new_len] == list('myurl')

# Пример 8: текст с двумя пробелами
text = list('  ')
new_len = urlify(text)
assert new_len == 6
assert text[:new_len] == list('%20%20')

print("Все тесты пройдены успешно!")
```
